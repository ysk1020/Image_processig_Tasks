```python
from google.colab import drive
drive.mount('/content/drive/')
import plotly.express as px
import imageio.v3 as iio
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import skimage as ski
```

```python
# Load your PNG image
img = iio.imread(uri='/content/drive/MyDrive/Colab Notebooks/images/Mokslininkai.png')

img_array = np.array(img[:,:,1])

# Display with Plotly (interactive zoom/pan)
fig = px.imshow(img_array, color_continuous_scale='gray')

fig.update_layout(
   title="Interactive Image Viewer",
   xaxis_showticklabels=False,
   yaxis_showticklabels=False
)

fig.show()
```

```python
pixels = img_array.flatten()

hist, bins = np.histogram(pixels, bins=256, range=(0, 255))

fig = px.bar(
    x=bins[:-1],
    y=hist,
    labels= {'x': 'Pixel Value', 'y': 'Frequency'},
    title='Histogram of Pixel Values'
)

fig.show()
```

# Region of Interest (ROI).

```python
mask= np.zeros(img_array.shape, dtype=bool)
start = (0 ,123)
extent =  (255-0, 969-123)
```

```python
from skimage.draw import rectangle
rr, cc = rectangle(start, extent, shape=img_array.shape)
mask[rr, cc] = True
px.imshow(mask.astype(int), color_continuous_scale="gray").show()
```

```python
face_pixels = img_array[mask]

hist, bins = np.histogram(
    face_pixels,
    bins=256,
    range=(0, 255)
)

fig = px.bar(
    x=bins[:-1],
    y=hist,
    title="Histogram of Selected Image Region",
    labels={"x": "Pixel Intensity", "y": "Pixel Count"}
)

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

fig.add_trace(go.Histogram(
    x=face_pixels,
    nbinsx=256,
    name="Faces",
    opacity=0.6
))

fig.add_trace(go.Histogram(
    x=pixels,
    nbinsx=256,
    name="Background",
    opacity=0.6
))

fig.update_layout(
    title="Histogram Comparison",
    xaxis_title="Pixel Intensity",
    yaxis_title="Pixel Count",
    barmode="overlay"
)

fig.show()
```
