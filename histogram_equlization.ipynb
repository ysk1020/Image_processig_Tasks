```python
from google.colab import drive
drive.mount('/content/drive/')
import plotly.express as px
import imageio.v3 as iio
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import skimage as ski
```

```python
# Load your PNG image

img = iio.imread(uri='/content/drive/MyDrive/Colab Notebooks/images/low_contrast.jpg')

img_array = np.array(img[:,:,1])

# hitogramos sudarymas

histogram, bin_edges = np.histogram(img_array, bins=256, range=(0, 256))

# Atvaizduojame histogramą
fig, ax = plt.subplots(1,2, figsize=(12,6))
ax[0].set_title("Image")
ax[0].imshow(img_array, cmap="gray")
ax[1].set_title("Grayscale Histogram")
ax[1].set_xlim([0, 256])
ax[1].bar(bin_edges[0:-1], histogram)
```

```python
# Equalize histogram
img_equalized = ski.exposure.equalize_hist(img_array)
img_equalized = (img_equalized * 255).astype(np.uint8)

# Histograms
hist_orig, bins_orig = np.histogram(img_array, bins=256, range=(0,256))
hist_eq, bins_eq = np.histogram(img_equalized, bins=256, range=(0,256))

# Subplots
fig, ax = plt.subplots(2, 2, figsize=(12, 12))

# Original image
ax[0,0].set_title("Original Image")
ax[0,0].imshow(img_array, cmap="gray")
ax[0,0].axis("off")

# Original histogram
ax[0,1].set_title("Original Histogram")
ax[0,1].bar(bins_orig[:-1], hist_orig)
ax[0,1].set_xlim([0,256])

# Equalized image
ax[1,0].set_title("Equalized Image")
ax[1,0].imshow(img_equalized, cmap="gray")
ax[1,0].axis("off")

# Equalized histogram
ax[1,1].set_title("Equalized Histogram")
ax[1,1].bar(bins_eq[:-1], hist_eq)
ax[1,1].set_xlim([0,256])

plt.tight_layout()
plt.show()
```

```python
# CLAHE
img_clahe = ski.exposure.equalize_adapthist(img_array, clip_limit=0.03)
img_clahe_uint8 = (img_clahe * 255).astype(np.uint8)

# Histograms (use uint8 for 0..255 bins)
hist_eq, bins_eq = np.histogram(img_equalized, bins=256, range=(0, 256))
hist_clahe, bins_clahe = np.histogram(img_clahe_uint8, bins=256, range=(0, 256))

# Plot
fig, ax = plt.subplots(2, 2, figsize=(12, 12))

# Equalized image
ax[0,0].set_title("Histogram Equalization")
ax[0,0].imshow(img_equalized, cmap="gray")
ax[0,0].axis("off")

# Equalized histogram
ax[0,1].set_title("Equalized Histogram")
ax[0,1].bar(bins_eq[:-1], hist_eq)
ax[0,1].set_xlim([0, 256])

# CLAHE image
ax[1,0].set_title("CLAHE")
ax[1,0].imshow(img_clahe_uint8, cmap="gray")
ax[1,0].axis("off")

# CLAHE histogram
ax[1,1].set_title("CLAHE Histogram")
ax[1,1].bar(bins_clahe[:-1], hist_clahe)
ax[1,1].set_xlim([0, 256])

plt.tight_layout()
plt.show()
```
